### MatplotLib
It's numerical mathematical extension Numpy. 

```python
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Graph in 2D")
#plt.savefig('Test.png')
```

```python
y = x*x
```

```python
plt.plot(x,y,'ro--')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('Graph in 2D')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'go')
```

```python
x = np.arange(1,11)
y = 3*x + 5
plt.plot(x,y)
plt.title("Matplotlib Demo")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
```

```python
np.pi
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y,'r--')
plt.title('Sine wave')
```

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,6,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2, color = 'g')
plt.title('Bar Graph')
```

### Histograms

```python
a = np.array([22,8,75,43,5,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins = 20)
plt.title('histogram')
```

### Box Plot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data, vert = True, patch_artist= True);
```

### Pie Chart

```python
labels = 'python', 'c++', 'Ruby', 'C'
sizes = [215,130,245,210]
colors = ['gold','yellow','lightcoral','lightskyblue']
explode = (0.1,0.0,0,0)

plt.pie(sizes, explode=explode, labels = labels, colors= colors,autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```
