### Seaborn
Distribution plots
- distplot
- joinplt (Bi variant features)
- pairplot(Multi variant features)
- on IRIS dataset

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

### tip dependent feature and total_bill, sex, smoker,day are independent features - correlation with heatmap 
Here correlation helps us to find interrelationship between features on int, float values

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

### Pair plot
Pair plot also known as scatterplot, if features are more than two independent features

```python
sns.pairplot(df)
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue='smoker')
```

### Dist Plot 
- Helps us to find distribution of the column features

```python
sns.distplot(df['tip'])
```

### Categorical plots
- Boxplot
- violinplot
- countplot
- Bar plot


```python
sns.countplot('sex',data= df)
```

```python
sns.countplot('day',data=df)
```

```python
# Bar Plot - have to give x and y values
sns.barplot(y = 'total_bill', x= 'sex',data=df)
```

```python
# Box plot
sns.boxplot('sex','total_bill', data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data = df, palette = 'rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

### Violin plot
Helps us to see both the distribution of data in terms of Kernel density estimation and box plot

```python
sns.violinplot(x='total_bill', y='day',data = df,palette='rainbow')
```
